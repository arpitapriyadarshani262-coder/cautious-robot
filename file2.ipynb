```python
#3
products = {
    "apple": 50,
    "milk": 40,
    "bread": 30,
    "rice": 60
}

cart = {}

while True:
    item = input("Enter product name (or 'done'): ")

    if item == "done":
        break

    if item in products:
        qty = int(input("Enter quantity: "))
        cart[item] = qty
    else:
        print("Product not available")

total = 0

print("\n----- BILL -----")
for item, qty in cart.items():
    price = products[item] * qty
    total += price
    print(item, "x", qty, "=", price)

discount = float(input("Enter discount (%): "))
total = total - (total * discount / 100)

print("Total Amount:", total)
```

```python
#4
seats = ["O"] * 10   # O = available, X = booked

def show_seats():
    print("Seats:")
    for i in range(10):
        print(i, seats[i])

while True:
    show_seats()

    seat = int(input("Enter seat number to book (-1 to exit): "))

    if seat == -1:
        break

    if seats[seat] == "O":
        seats[seat] = "X"
        print("Seat booked successfully")
    else:
        print("Seat already booked")
```

```python
#2
tasks = []

def add_task():
    name = input("Enter task name: ")
    category = input("Enter category: ")
    priority = input("Enter priority (High/Medium/Low): ")
    deadline = input("Enter deadline: ")

    task = {
        "name": name,
        "category": category,
        "priority": priority,
        "deadline": deadline,
        "done": False
    }

    tasks.append(task)

def view_tasks():
    for i, task in enumerate(tasks):
        status = "Done" if task["done"] else "Pending"
        print(i, task["name"], "|", task["category"], "|", task["priority"], "|", task["deadline"], "|", status)

def complete_task():
    view_tasks()
    index = int(input("Enter task number to mark complete: "))
    tasks[index]["done"] = True

while True:
    print("\n1.Add Task\n2.View Tasks\n3.Mark Complete\n4.Exit")
    choice = input("Enter choice: ")

    if choice == "1":
        add_task()
    elif choice == "2":
        view_tasks()
    elif choice == "3":
        complete_task()
    elif choice == "4":
        break
```

```python
#5
students = {}

def add_student():
    name = input("Enter student name: ")
    marks = []

    for i in range(3):
        m = int(input(f"Enter marks for subject {i+1}: "))
        marks.append(m)

    avg = sum(marks) / len(marks)

    students[name] = {
        "marks": marks,
        "GPA": avg / 10
    }

def show_results():
    for name, data in students.items():
        print("\nName:", name)
        print("Marks:", data["marks"])
        print("GPA:", round(data["GPA"],2))

while True:
    print("\n1.Add Student\n2.View Results\n3.Exit")
    choice = input("Enter choice: ")

    if choice == "1":
        add_student()
    elif choice == "2":
        show_results()
    elif choice == "3":
        break
```

```python
#1
class ParkingLot:
    def __init__(self, total_spots):
        self.total_spots = total_spots
        self.available_spots = total_spots
        self.vehicles = {}

    def entry(self, vehicle_number):
        if self.available_spots > 0:
            self.vehicles[vehicle_number] = 0
            self.available_spots -= 1
            print("Vehicle parked successfully.")
        else:
            print("Parking Full!")

    def exit(self, vehicle_number, hours):
        if vehicle_number in self.vehicles:
            fee = hours * 20
            self.available_spots += 1
            del self.vehicles[vehicle_number]
            print("Vehicle exited. Parking Fee =", fee)
        else:
            print("Vehicle not found.")

    def status(self):
        print("Available Spots:", self.available_spots)


# Example
p = ParkingLot(5)
p.entry("OD02AB1234")
p.exit("OD02AB1234", 3)
p.status()
```
